```python
def plot_single(list_fn, tag):
    for fn in list_fn:
        plt.plot(x, fn[0], label=fn[1])
    plt.legend()
    plt.grid(True)
    plt.show()
    
#     fig, ax = plt.subplots(figsize=(16,10))
#     fig.suptitle(tag)
```

```python
def plot_multiple(list_fn, columns=2, tag="test"):
    rows = int(len(list_fn) / columns) + int(len(list_fn) % columns != 0)
    fs = (columns * 10, rows * 7)
#     dprint(fs)
    fig, axs = plt.subplots(rows, columns, figsize=fs, squeeze=False)
    fig.suptitle(tag)

    for i, fn in enumerate(list_fn):
        j = int(i / columns)
        k = i % columns
        axs[j][k].plot(x, fn[0], label=fn[1], linewidth=1)
        axs[j][k].legend()
        axs[j][k].grid(True)
    plt.show()
```

```python
def plots(list_fn_str, multiple=True, columns=2, tag="test"):
    Y = []
    for fn_str in list_fn_str:
        Y.append((eval(fn_str), fn_str))
    if multiple:
        plot_multiple(Y, columns, tag)
    else:
        plot_single(Y, tag)
```
